```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data of the first model evaluation
data_1 = {
    'Difficulty': ['Easy', 'Medium', 'Hard', 'Extra', 'All'],
    'Execution Accuracy': [0.742, 0.666, 0.517, 0.361, 0.610],
    'Exact Matching Accuracy': [0.690, 0.619, 0.454, 0.367, 0.568],
    'Model': ['Spider Skeleton Wizard Coder']*5
}

# Data of the second model evaluation
data_2 = {
    'Difficulty': ['Easy', 'Medium', 'Hard', 'Extra', 'All'],
    'Execution Accuracy': [0.750, 0.648, 0.460, 0.247, 0.576],
    'Exact Matching Accuracy': [0.464, 0.265, 0.167, 0.042, 0.260],
    'Model': ['ChatGPT']*5
}

# Create DataFrame
df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)

# Concatenate the two DataFrames
df = pd.concat([df_1, df_2])

# Melt the DataFrame
df_melted = df.melt(['Difficulty', 'Model'], var_name='Metrics', value_name='Values')

# Set style
sns.set_style("whitegrid")

# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 2, figsize=(22, 8))

pallette = {'Spider Skeleton Wizard Coder': 'royalblue', 'ChatGPT': 'darkorange'}

# Loop over each metric
for ax, metric in zip(axes, ['Execution Accuracy', 'Exact Matching Accuracy']):
    # Filter data for the current metric
    df_metric = df_melted[df_melted['Metrics'] == metric]
    
    # Barplot for each model
    sns.barplot(x='Difficulty', y='Values', hue='Model', data=df_metric, ax=ax, palette=pallette)
    
    # Set plot details
    ax.set_title(metric, fontsize=16)
    ax.set_xlabel('Difficulty', fontsize=14)
    ax.set_ylabel('Values', fontsize=14)
    ax.set_ylim(0, 1)  # Setting y limit from 0 to 1

# Overall title
fig.suptitle('Comparison of Spider Skeleton Wizard Coder and ChatGPT', fontsize=20)

# Save plot
plt.savefig('spiderwizard-vs-chatgpt.svg')

# Show the plot
plt.show()
```
